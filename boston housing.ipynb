```python
import pandas as pd
import numpy as np
import seaborn as ans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Boston Dataset.csv")
```

```python
df.head()
```

```python
df.drop(columns=["Unnamed: 0"],axis=0, inplace = True)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df_y = df.drop(['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'black', 'lstat'],axis = 1)
```

```python
df_y.reset_index
```

```python
df_x= df.drop(['medv'],axis=1)
```

```python
df_x
```

```python
df_x.describe()
```

```python
from sklearn import linear_model
```

```python
from sklearn.model_selection import train_test_split
```

```python
model = linear_model.LinearRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_x,df_y,test_size=0.22,random_state=42)
```

```python
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
print(y_test)
```

```python
print(np.mean((y_pred - y_test)**2))
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,y_pred)
```

```python
model.score(X_test,y_test)
```
